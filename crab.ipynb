```python
import pandas as pd

# Load the dataset
file_path = 'crab_dataset.xlsx'  # Replace with your file path
df_2018 = pd.read_excel(file_path, sheet_name='Lobster Data 2018')
df_2019 = pd.read_excel(file_path, sheet_name='Lobster Data 2019')

# Display the first few rows of each sheet
print("2018 Data:")
print(df_2018.head())
print("\n2019 Data:")
print(df_2019.head())
```

```python
df_2018 = df_2018.fillna(0)
df_2018.head()
```

```python
df_2019 = df_2019.fillna(0)
df_2019.head()
```

```python
print(df_2018.isna().sum())  # Count of NaN values per column
print((df_2018 == 0).sum())  # Count of zero values per column
print(df_2019.isna().sum())  # Count of NaN values per column
print((df_2019 == 0).sum())  # Count of zero values per column
```

```python
import pandas as pd
import numpy as np

# Calculate mean values by Sex
mean_values = df_2018.groupby('Sex').mean()

# Function to replace zeros with mean values
def replace_zeros_with_mean(row):
    if row['Sex'] in mean_values.index:
        for col in ['Length(mm)', 'Diameter(mm)', 'Height(mm)', 'WholeWeight(g)', 'ShuckedWeight(g)', 'SellWeight(g)']:
            if row[col] == 0:
                row[col] = mean_values.loc[row['Sex'], col]
    return row

# Apply the function to replace zeros
df_2018_cleaned = df_2018.apply(replace_zeros_with_mean, axis=1)

print("Original DataFrame:")
print(df_2018)

print("\nDataFrame with zeros replaced by means:")
print(df_2018_cleaned)
```

```python
import pandas as pd
import numpy as np

# Calculate mean values by Sex
mean_values = df_2019.groupby('Sex').mean()

# Function to replace zeros with mean values
def replace_zeros_with_mean(row):
    if row['Sex'] in mean_values.index:
        for col in ['Length(mm)', 'Diameter(mm)', 'Height(mm)', 'WholeWeight(g)', 'ShuckedWeight(g)', 'SellWeight(g)']:
            if row[col] == 0:
                row[col] = mean_values.loc[row['Sex'], col]
    return row

# Apply the function to replace zeros
df_2019_cleaned = df_2019.apply(replace_zeros_with_mean, axis=1)

print("Original DataFrame:")
print(df_2019)

print("\nDataFrame with zeros replaced by means:")
print(df_2019_cleaned)
```

```python
import pandas as pd

# Load the existing cleaned data from the Excel file
input_file = 'modified_crab_dataset.xlsx'
df_2018_cleaned = pd.read_excel(input_file, sheet_name='Lobster 2018')
df_2019_cleaned = pd.read_excel(input_file, sheet_name='Lobster 2019')

# Add a column for the year
df_2018_cleaned['Year'] = 2018
df_2019_cleaned['Year'] = 2019

# Concatenate the DataFrames into a single DataFrame
combined_df = pd.concat([df_2018_cleaned, df_2019_cleaned], ignore_index=True)

# Specify the output file path
output_file_combined = 'combined_lobster_data.xlsx'

# Export the combined DataFrame to a new Excel file
combined_df.to_excel(output_file_combined, index=False)

print(f"Combined data exported to {output_file_combined} successfully.")
```
